```python
# Import libraries
import pandas as pd
import re
import nltk

nltk.download('stopwords')
nltk.download('punkt')
```

```python
df_comments = pd.read_csv('../Copy of Youtube Analysis/df_comments.csv')
df_videos = pd.read_csv('../Copy of Youtube Analysis/df_videos.csv')
```

# Comments

```python
df_comments.info()
df_comments.head()
```

```python
# Load the stopwords
stopwords = nltk.corpus.stopwords.words('english')

# Define a function to preprocess the text
def preprocess_text(text):
    # Convert the text to lowercase
    text = text.lower()

     # Remove URLs
    text = re.sub(r'http\S+', '', str(text))

    # Remove usernames
    text = re.sub(r'@\w+', '', str(text))

    # Remove hashtags
    text = re.sub(r'#\w+', '', str(text))

    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', str(text))

    # Remove emoticons
    text = re.sub(r':\)|;\)|:-\)|\(-:|:\'\(|:\(|:-\(|:O|8\)|:\*|>:O|:\-\/|\^\^|:D|:P|:S|:\|', '', str(text))

    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', str(text)).strip()
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove stopwords
    text = [word for word in text.split() if word not in stopwords]

    # Detokenize the tokens to form text
    text = ' '.join(text)

    return text
```

```python
# Apply the preprocess_text() function on text column
df_comments['preprocessed_text'] = df_comments['text'].apply(lambda x: preprocess_text(x))
df_comments.info()
df_comments.head()
```
