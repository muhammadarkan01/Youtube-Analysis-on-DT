# **2. Sentiment Labelling**

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
from tqdm import tqdm
```

```python
# Instantiate pre-trained model
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment', clean_up_tokenization_spaces=True)
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
# Define function to execute the pre-trained model
def sentiment_score(text):
    tokens = tokenizer.encode(text, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
# # Slicing dataset
# new_df = df_comments[:6000]
# new_df.info()
# new_df.head()
```

```python
# # Execute sentiment labelling
# tqdm.pandas(desc='My bar!')
# new_df['sentiment_score'] = new_df['preprocessed_text'].progress_apply(lambda x: sentiment_score(x[:512]))

# Execute sentiment labelling
tqdm.pandas(desc='My bar!')
df_comments['sentiment_score'] = df_comments['preprocessed_text'].progress_apply(lambda x: sentiment_score(x[:512]))
```

```python
# Execute sentiment labelling
tqdm.pandas(desc='My bar!')
df_comments['sentiment_score_2'] = df_comments['text'].progress_apply(lambda x: sentiment_score(x[:512]))
```

```python
df_comments
```

```python
# Define the function to categorize sentiment scores
def categorize_sentiment(row):
    score1 = row['sentiment_score']
    score2 = row['sentiment_score_2']
    
    # Calculate the average of the two sentiment scores
    average_score = (score1 + score2) / 2
    
    # Categorize based on the average score
    if average_score < 2.5:
        return 'negative'
    elif 2.5 <= average_score <= 3.5:
        return 'neutral'
    else:
        return 'positive'

# Apply the function to create a new column in the DataFrame
df_comments['sentiment_category'] = df_comments.apply(categorize_sentiment, axis=1)
df_comments
```

```python
# Count occurrences of each category
category_counts = df_comments['sentiment_category'].value_counts()
category_counts
```
