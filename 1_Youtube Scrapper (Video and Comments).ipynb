```python
import glob
import json
import pandas as pd

# Find all files with extension ".info.json" in the current directory
file_paths = glob.glob("*.info.json")

df_videos = pd.DataFrame()
df_comments = pd.DataFrame()

# Read and process each file
for file_path in file_paths:
    with open(file_path, "r", encoding='utf-8') as file:
        data = json.load(file)

        df_videos_record = pd.DataFrame([{
            'id': data['id'],
            'title': data['title'],
            'duration': data['duration'],
            'view_count': data['view_count'],
            'webpage_url': data['webpage_url'],
            'categories': data['categories'],
            'comment_count': data.get('comment_count', 0),
            'like_count': data['like_count'],
            'tags': data['tags'],
            'upload_date': data['upload_date']
        }])

        df_videos = pd.concat([df_videos, df_videos_record], ignore_index=True)

        if 'comment_count' in data.keys():
            for comment in data['comments']:
                df_comment_record = pd.DataFrame([{
                    'id': comment['id'],
                    'video_id': data['id'],
                    'text': comment['text'],
                    'timestamp': comment['timestamp'],
                    'author': comment.get('author', None),  # Set to None if 'author' doesn't exist
                    'parent': comment['parent']
                }])

                df_comments = pd.concat([df_comments, df_comment_record], ignore_index=True)

# Display the resulting DataFrames
print("Videos DataFrame:")
print(df_videos.head())

print("\nComments DataFrame:")
print(df_comments.head())
```
